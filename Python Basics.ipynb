````python
#Write a python program to convert user input from string to integer float and boolean type.
try:
  user_value = input("Enter a value:")
  int_value = int(user_value)
  print("Integer:", int_value)

except ValueError:
  print("Cannot convert to Integer")

try:
  float_value = float(user_value)
  print("Float:", float_value)

except ValueError:
  print("Cannot convert to Float")

bool_value = bool(user_value)
print("Boolean:", bool_value)
````

````python
#Write code to demonstrate type casting with listt elements.
data = ["10", "20.5","True","hello"]
int_list = [int(x)for x in data if x.isdigit()]
print("Integer List:", int_list)

float_list = [float(x)for x in data if x.replace(".","",1).isdigit()]
print("Float List:", float_list)

bool_list = [bool(x)for x in data]
print("Boolean List:", bool_list)

tuple_data = tuple(data)
print("Tuple:", tuple_data)
````

````python
#Write a program to reverse a string using a while loop.
text = input ("Enter a string:")
reversed_text = ""
index = len(text) - 1
while index > 0:
  reversed_text += text[index]
  print("Reversed string:", reversed_text)
````

````python
#Write a python program to calculate the factorial of a number provide by the user using a while loop
num = int(input("Enter a number:"))
factorial = 1
i = num
while i > 0:
  factorial *=i
  i -=1
  print(f"factorial of {num} is {factorial}")
````
